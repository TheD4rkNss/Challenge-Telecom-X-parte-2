# Modelagem Preditiva

### Separação dos Dados

```python
from sklearn.model_selection import train_test_split
```

```python
# Separar features e target
X = df_encoded.drop('Churn', axis = 1)
y = df_encoded['Churn']

# Divisão estratificada para manter a proporção de churn

X_escala = X.copy()
X_train, X_test, y_train, y_test = train_test_split(
    X_escala, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

print(f"Shape X_train: {X_train.shape}")
print(f"Shape X_test: {X_test.shape}")
print(f"Shape y_train: {y_train.shape}")
print(f"Shape y_test: {y_test.shape}")

print(f"\nDistribuição do target no conjunto de treino (70%):")
print(y_train.value_counts(normalize=True))

print(f"\nDistribuição do target no conjunto de teste (30%):")
print(y_test.value_counts(normalize=True))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
# ================================
# Modelo 1: Regressão Logística (exige normalização)
# ================================
print("\nModelo 1: Regressão Logística")
print("-"*50)
print("Justificativa: A Regressão Logística é um modelo linear amplamente utilizado para classificação binária, como o problema de churn. Ela é sensível à escala das variáveis, por isso a normalização é importante para garantir que todas as features contribuam igualmente para o modelo.")

# Treinando o modelo de Regressão Logística
logreg = LogisticRegression(random_state=42, max_iter=1000)
logreg.fit(X_train, y_train)

# Avaliação no conjunto de teste
y_pred_logreg = logreg.predict(X_test)
y_proba_logreg = logreg.predict_proba(X_test)[:,1]

print("Acurácia:", accuracy_score(y_test, y_pred_logreg))
print("ROC AUC:", roc_auc_score(y_test, y_proba_logreg))
print("\nClassification Report:\n", classification_report(y_test, y_pred_logreg))

# Matriz de confusão
plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test, y_pred_logreg), annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusão - Regressão Logística')
plt.xlabel('Predito')
plt.ylabel('Real')
plt.show()

# Curva ROC
fpr, tpr, _ = roc_curve(y_test, y_proba_logreg)
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc_score(y_test, y_proba_logreg):.2f})')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('Falso Positivo')
plt.ylabel('Verdadeiro Positivo')
plt.title('Curva ROC - Regressão Logística')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ================================
# Modelo 2: Random Forest (não exige normalização)
# ================================
print("\nModelo 2: Random Forest")
print("-"*50)
print("Justificativa: O Random Forest é um modelo de ensemble baseado em árvores de decisão. Ele não é sensível à escala das variáveis, pois as árvores utilizam apenas ordenação dos valores para realizar splits. Portanto, a normalização não é obrigatória para esse tipo de modelo.")

# Para demonstrar que a normalização não é necessária, vamos treinar o Random Forest com os dados originais (sem normalização)
# Recriando X_train_rf e X_test_rf a partir de X (antes do StandardScaler)
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_rf, y_train_rf)

y_pred_rf = rf.predict(X_test_rf)
y_proba_rf = rf.predict_proba(X_test_rf)[:,1]

print("Acurácia:", accuracy_score(y_test_rf, y_pred_rf))
print("ROC AUC:", roc_auc_score(y_test_rf, y_proba_rf))
print("\nClassification Report:\n", classification_report(y_test_rf, y_pred_rf))

# Matriz de confusão
plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test_rf, y_pred_rf), annot=True, fmt='d', cmap='Greens')
plt.title('Matriz de Confusão - Random Forest')
plt.xlabel('Predito')
plt.ylabel('Real')
plt.show()

# Curva ROC
fpr_rf, tpr_rf, _ = roc_curve(y_test_rf, y_proba_rf)
plt.figure(figsize=(6,5))
plt.plot(fpr_rf, tpr_rf, label=f'ROC curve (AUC = {roc_auc_score(y_test_rf, y_proba_rf):.2f})', color='green')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('Falso Positivo')
plt.ylabel('Verdadeiro Positivo')
plt.title('Curva ROC - Random Forest')
plt.legend()
plt.tight_layout()
plt.show()

# ================================
# COMPARATIVO FINAL
# ================================
print("\n" + "="*50)
print("COMPARATIVO FINAL ENTRE OS MODELOS")
print("="*50)
print("Resumo das justificativas:")
print("- Regressão Logística: requer normalização pois é sensível à escala das variáveis. Ideal para interpretar pesos das features e obter probabilidades.")
print("- Random Forest: não requer normalização, pois utiliza apenas a ordenação dos valores. É robusto a outliers e pode capturar relações não-lineares.")

print("\nResultados de Acurácia e ROC AUC:")
print(f"Regressão Logística - Acurácia: {accuracy_score(y_test, y_pred_logreg):.3f} | ROC AUC: {roc_auc_score(y_test, y_proba_logreg):.3f}")
print(f"Random Forest        - Acurácia: {accuracy_score(y_test_rf, y_pred_rf):.3f} | ROC AUC: {roc_auc_score(y_test_rf, y_proba_rf):.3f}")

print("\nAmbos os modelos podem ser utilizados para prever a evasão de clientes, sendo a escolha final dependente do objetivo (interpretação vs performance")
```

```python
print("\n" + "="*50)
print("AVALIAÇÃO DETALHADA DOS MODELOS")
print("="*50)

# Avaliação para Regressão Logística
print("\nMétricas - Regressão Logística")
print("-"*50)
acc_logreg = accuracy_score(y_test, y_pred_logreg)
prec_logreg = precision_score(y_test, y_pred_logreg)
rec_logreg = recall_score(y_test, y_pred_logreg)
f1_logreg = f1_score(y_test, y_pred_logreg)
cm_logreg = confusion_matrix(y_test, y_pred_logreg)

print(f"Acurácia : {acc_logreg:.3f}")
print(f"Precisão : {prec_logreg:.3f}")
print(f"Recall   : {rec_logreg:.3f}")
print(f"F1-score : {f1_logreg:.3f}")
print("Matriz de Confusão:")
print(cm_logreg)

# Avaliação para Random Forest
print("\nMétricas - Random Forest")
print("-"*50)
acc_rf = accuracy_score(y_test_rf, y_pred_rf)
prec_rf = precision_score(y_test_rf, y_pred_rf)
rec_rf = recall_score(y_test_rf, y_pred_rf)
f1_rf = f1_score(y_test_rf, y_pred_rf)
cm_rf = confusion_matrix(y_test_rf, y_pred_rf)

print(f"Acurácia : {acc_rf:.3f}")
print(f"Precisão : {prec_rf:.3f}")
print(f"Recall   : {rec_rf:.3f}")
print(f"F1-score : {f1_rf:.3f}")
print("Matriz de Confusão:")
print(cm_rf)
```

```python
# Comparação de desempenho
print("\nComparação de desempenho dos modelos:")

print(f"Regressão Logística - Acurácia: {acc_logreg:.3f} | Precisão: {prec_logreg:.3f} | Recall: {rec_logreg:.3f} | F1-score: {f1_logreg:.3f} | ROC AUC: {roc_auc_score(y_test, y_proba_logreg):.3f}")
print(f"Random Forest        - Acurácia: {acc_rf:.3f} | Precisão: {prec_rf:.3f} | Recall: {rec_rf:.3f} | F1-score: {f1_rf:.3f} | ROC AUC: {roc_auc_score(y_test_rf, y_proba_rf):.3f}")

# Determinar qual modelo teve melhor desempenho
if roc_auc_score(y_test, y_proba_logreg) > roc_auc_score(y_test_rf, y_proba_rf):
    melhor_modelo = "Regressão Logística"
else:
    melhor_modelo = "Random Forest"

print(f"\n➡️  O modelo com melhor desempenho geral (considerando ROC AUC) foi: {melhor_modelo}")

# Análise de overfitting/underfitting
print("\nAnálise de Overfitting e Underfitting:")

# Avaliação do desempenho em treino e teste para ambos os modelos
# Para Logística
train_pred_logreg = logreg.predict(X_train)
train_proba_logreg = logreg.predict_proba(X_train)[:,1]
acc_train_logreg = accuracy_score(y_train, train_pred_logreg)
roc_train_logreg = roc_auc_score(y_train, train_proba_logreg)

# Para Random Forest
train_pred_rf = rf.predict(X_train_rf)
train_proba_rf = rf.predict_proba(X_train_rf)[:,1]
acc_train_rf = accuracy_score(y_train_rf, train_pred_rf)
roc_train_rf = roc_auc_score(y_train_rf, train_proba_rf)

print(f"\nRegressão Logística - Treino: Acurácia: {acc_train_logreg:.3f} | ROC AUC: {roc_train_logreg:.3f}")
print(f"Regressão Logística - Teste : Acurácia: {acc_logreg:.3f} | ROC AUC: {roc_auc_score(y_test, y_proba_logreg):.3f}")

print(f"\nRandom Forest        - Treino: Acurácia: {acc_train_rf:.3f} | ROC AUC: {roc_train_rf:.3f}")
print(f"Random Forest        - Teste : Acurácia: {acc_rf:.3f} | ROC AUC: {roc_auc_score(y_test_rf, y_proba_rf):.3f}")

# Diagnóstico
# Overfitting: desempenho muito maior em treino do que em teste
# Underfitting: desempenho baixo em ambos

# Análise para Regressão Logística
if (acc_train_logreg - acc_logreg) > 0.05 or (roc_train_logreg - roc_auc_score(y_test, y_proba_logreg)) > 0.05:
    print("\n➡️  A Regressão Logística apresenta sinais de overfitting (desempenho significativamente maior no treino do que no teste).")
    print("Possíveis causas: modelo pode estar capturando ruídos dos dados, ou há muitas variáveis irrelevantes.")
    print("Ajustes sugeridos: regularização (ajustar parâmetro C), seleção de features, ou coleta de mais dados.")
elif acc_logreg < 0.7 and acc_train_logreg < 0.7:
    print("\n➡️  A Regressão Logística apresenta sinais de underfitting (desempenho baixo em ambos os conjuntos).")
    print("Possíveis causas: modelo muito simples, falta de variáveis relevantes ou relações não-lineares não capturadas.")
    print("Ajustes sugeridos: adicionar features, tentar modelos mais complexos, ou transformar variáveis.")
else:
    print("\n➡️  A Regressão Logística apresenta desempenho consistente entre treino e teste, sem sinais claros de overfitting ou underfitting.")

# Análise para Random Forest
if (acc_train_rf - acc_rf) > 0.10 or (roc_train_rf - roc_auc_score(y_test_rf, y_proba_rf)) > 0.10:
    print("\n➡️  O Random Forest apresenta sinais de overfitting (desempenho muito alto no treino comparado ao teste).")
    print("Possíveis causas: modelo muito complexo (muitas árvores ou profundidade alta), dados ruidosos.")
    print("Ajustes sugeridos: reduzir a profundidade máxima das árvores (max_depth), aumentar min_samples_leaf, ou usar mais regularização.")
elif acc_rf < 0.7 and acc_train_rf < 0.7:
    print("\n➡️  O Random Forest apresenta sinais de underfitting (desempenho baixo em ambos os conjuntos).")
    print("Possíveis causas: modelo pouco complexo, poucas árvores, ou falta de variáveis relevantes.")
    print("Ajustes sugeridos: aumentar o número de árvores (n_estimators), adicionar features, ou ajustar hiperparâmetros.")
else:
    print("\n➡️  O Random Forest apresenta desempenho consistente entre treino e teste, sem sinais claros de overfitting ou underfitting.")

print("\nResumo:")
print(f"- O modelo com melhor desempenho foi: {melhor_modelo}")
print("- Overfitting ocorre quando o modelo aprende demais os dados de treino e perde generalização; underfitting ocorre quando o modelo não consegue capturar padrões relevantes.")
print("- Ajustes de regularização, seleção de variáveis e tuning de hiperparâmetros podem melhorar o equilíbrio entre viés e variância dos modelos.")
```

```python
print("\n" + "="*50)
print("AJUSTE DE RANDOM FOREST PARA REDUZIR OVERFITTING")
print("="*50)

from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import GridSearchCV

# 1. Seleção de variáveis baseada em importância
print("\nSelecionando features mais importantes com Random Forest...")

# Treinar um Random Forest simples para obter importâncias
rf_base = RandomForestClassifier(n_estimators=100, random_state=42)
rf_base.fit(X_train_rf, y_train_rf)

importances = rf_base.feature_importances_
feature_names = X_train_rf.columns
feat_imp_df = pd.DataFrame({'feature': feature_names, 'importance': importances})
feat_imp_df = feat_imp_df.sort_values(by='importance', ascending=False)

# Selecionar as top 20 features (ou todas se menos de 20)
top_n = min(20, len(feature_names))
top_features = feat_imp_df['feature'].head(top_n).tolist()
print(f"Top {top_n} features selecionadas para o ajuste: {top_features}")

X_train_sel = X_train_rf[top_features]
X_test_sel = X_test_rf[top_features]

# 2. Tuning de hiperparâmetros com GridSearchCV
print("\nRealizando tuning de hiperparâmetros com GridSearchCV...")

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [4, 6, 8, None],
    'min_samples_leaf': [1, 2, 4, 8],
    'max_features': ['sqrt', 'log2', None]
}

rf_grid = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(
    rf_grid, 
    param_grid, 
    cv=3, 
    scoring='roc_auc', 
    n_jobs=-1,
    verbose=1
)
grid_search.fit(X_train_sel, y_train_rf)

print(f"\nMelhores hiperparâmetros encontrados: {grid_search.best_params_}")

# 3. Avaliação do modelo ajustado
rf_best = grid_search.best_estimator_
y_pred_rf_best = rf_best.predict(X_test_sel)
y_proba_rf_best = rf_best.predict_proba(X_test_sel)[:,1]

print("\nDesempenho do Random Forest ajustado:")
print(f"Acurácia: {accuracy_score(y_test_rf, y_pred_rf_best):.3f}")
print(f"ROC AUC: {roc_auc_score(y_test_rf, y_proba_rf_best):.3f}")
print("\nClassification Report:\n", classification_report(y_test_rf, y_pred_rf_best))

# Matriz de confusão
plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test_rf, y_pred_rf_best), annot=True, fmt='d', cmap='Purples')
plt.title('Matriz de Confusão - Random Forest Ajustado')
plt.xlabel('Predito')
plt.ylabel('Real')
plt.show()

# Curva ROC
fpr_best, tpr_best, _ = roc_curve(y_test_rf, y_proba_rf_best)
plt.figure(figsize=(6,5))
plt.plot(fpr_best, tpr_best, label=f'ROC curve (AUC = {roc_auc_score(y_test_rf, y_proba_rf_best):.2f})', color='purple')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('Falso Positivo')
plt.ylabel('Verdadeiro Positivo')
plt.title('Curva ROC - Random Forest Ajustado')
plt.legend()
plt.tight_layout()
plt.show()

# 4. Comparação com o modelo anterior
print("\nComparação de desempenho antes e depois do ajuste:")
print(f"Random Forest Original - Teste: Acurácia: {acc_rf:.3f} | ROC AUC: {roc_auc_score(y_test_rf, y_proba_rf):.3f}")
print(f"Random Forest Ajustado - Teste: Acurácia: {accuracy_score(y_test_rf, y_pred_rf_best):.3f} | ROC AUC: {roc_auc_score(y_test_rf, y_proba_rf_best):.3f}")

# 5. Avaliação de overfitting após ajuste
train_pred_rf_best = rf_best.predict(X_train_sel)
train_proba_rf_best = rf_best.predict_proba(X_train_sel)[:,1]
acc_train_rf_best = accuracy_score(y_train_rf, train_pred_rf_best)
roc_train_rf_best = roc_auc_score(y_train_rf, train_proba_rf_best)

print(f"\nRandom Forest Ajustado - Treino: Acurácia: {acc_train_rf_best:.3f} | ROC AUC: {roc_train_rf_best:.3f}")
print(f"Random Forest Ajustado - Teste : Acurácia: {accuracy_score(y_test_rf, y_pred_rf_best):.3f} | ROC AUC: {roc_auc_score(y_test_rf, y_proba_rf_best):.3f}")

if (acc_train_rf_best - accuracy_score(y_test_rf, y_pred_rf_best)) > 0.10 or (roc_train_rf_best - roc_auc_score(y_test_rf, y_proba_rf_best)) > 0.10:
    print("\n➡️  Ainda há sinais de overfitting. Considere ajustar ainda mais max_depth, min_samples_leaf ou usar mais regularização.")
else:
    print("\n➡️  O ajuste reduziu o overfitting e melhorou o equilíbrio entre viés e variância do Random Forest.")

print("\nResumo do ajuste:")
print("- Seleção de variáveis importantes (feature selection)")
print("- Ajuste de max_depth, min_samples_leaf e max_features")
print("- Tuning de hiperparâmetros com GridSearchCV")
print("- Avaliação do desempenho e comparação com o modelo original")
```

## 4. Estratégias de Retenção Propostas

Com base nos fatores identificados, sugerem-se as seguintes estratégias:

### 1. **Ações de Boas-Vindas e Engajamento Inicial**
   - Desenvolver programas de onboarding para novos clientes, focando nos primeiros meses de contrato.
   - Oferecer benefícios ou descontos progressivos para clientes que completam marcos de tempo (ex: 6 meses, 1 ano).

### 2. **Revisão de Preços e Planos**
   - Avaliar a competitividade das taxas mensais e cobranças totais, especialmente para clientes com maior risco de evasão.
   - Oferecer planos personalizados ou descontos para clientes que sinalizam insatisfação com o preço.

### 3. **Promoção de Serviços Adicionais**
   - Incentivar a adesão a serviços complementares (streaming, backup, proteção), mostrando o valor agregado.
   - Criar bundles promocionais para aumentar o engajamento e a fidelização.

### 4. **Foco em Clientes de Alto Valor**
   - Implementar programas de relacionamento exclusivos para clientes de alto valor, com atendimento diferenciado e ofertas especiais.
   - Monitorar sinais de insatisfação neste grupo e agir proativamente.

### 5. **Incentivo a Contratos de Longo Prazo**
   - Oferecer condições vantajosas para migração de contratos mensais para anuais ou bianuais.
   - Reduzir barreiras para renovação automática e facilitar o processo de upgrade de contrato.

### 6. **Facilitação e Segurança no Pagamento**
   - Promover métodos de pagamento automáticos e seguros, reduzindo o churn associado a pagamentos eletrônicos avulsos.
   - Oferecer suporte dedicado para resolução de problemas de cobrança.

### 7. **Aprimoramento do Suporte Técnico**
   - Investir em canais de suporte eficientes e proativos, especialmente para clientes que não utilizam serviços de suporte ou segurança.
   - Realizar pesquisas de satisfação após interações de suporte para identificar pontos de melhoria.


## 5. Conclusão

A análise dos dados e dos modelos evidencia que a evasão de clientes é multifatorial, envolvendo aspectos de tempo de relacionamento, engajamento com serviços, sensibilidade a preço e experiência de suporte. A adoção das estratégias propostas pode contribuir significativamente para a redução do churn e o aumento da fidelização dos clientes da Telecom X.
