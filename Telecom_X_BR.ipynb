# Preparação os dados para a modelagem

```python
df = df.drop(columns = ['Id_cliente'], axis= 1) #Removendo a coluna Id_clinte, pois não contribui para nosso modelo
```

```python
df.info() # Verificando se a coluna foi excluida
```

```python
col_categoricas = df.select_dtypes(include=['object']).columns
col_categoricas
```

```python
df_encoded = pd.get_dummies(df, columns = col_categoricas, drop_first= True)
df_encoded.head()
```

```python
df['Churn'].value_counts(normalize=True) * 100
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
escala = StandardScaler()
col_numericas = ['Taxas_mensais', 'Cobranca_total']
df_encoded[col_numericas] = escala.fit_transform(df_encoded[col_numericas])
df_encoded.head()
```
