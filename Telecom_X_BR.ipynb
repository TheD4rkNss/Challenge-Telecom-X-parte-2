# Preparação os dados para a modelagem

```python
df = df.drop(columns = ['Id_cliente'], axis= 1) #Removendo a coluna Id_clinte, pois não contribui para nosso modelo
```

```python
df.info() # Verificando se a coluna foi excluida
```

```python
col_categoricas = df.select_dtypes(include=['object']).columns
col_categoricas
```

```python
df_encoded = pd.get_dummies(df, columns = col_categoricas, drop_first= True)
df_encoded.head()
```

```python
# Verifica a proporção de churn
df['Churn'].value_counts(normalize=True) * 100
```

```python
# visualização Grafica
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,6))
churn_counts = df['Churn'].value_counts().sort_index()
sns.barplot(x=churn_counts.index, y=churn_counts.values, palette='husl', hue = churn_counts.index)
plt.title('Quantidade de Clientes Inadimplentes (Churn) vs Não Inadimplentes')
plt.xlabel('Churn (0 = Não Inadimplente, 1 = Inadimplente)')
plt.ylabel('Quantidade de Clientes')
plt.xticks([0,1], ['Não Inadimplente', 'Inadimplente'])
for i, v in enumerate(churn_counts.values):
    plt.text(i, v + max(churn_counts.values)*0.01, str(v), ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

# Análise de Correlação

```python
# VISUALIZAÇÃO DA MATRIZ DE CORRELAÇÃO DAS VARIÁVEIS NUMÉRICAS

# Selecionar apenas as variáveis numéricas do dataframe original tratado
numeric_cols_corr = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
if 'Churn' in numeric_cols_corr:
    numeric_cols_corr.remove('Churn')

# Adicionar a variável target para análise de correlação
corr_df = df[numeric_cols_corr + ['Churn']]

# Calcular a matriz de correlação
corr_matrix = corr_df.corr()

# Visualizar a matriz de correlação
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Matriz de Correlação - Variáveis Numéricas')
plt.tight_layout()
plt.show()
```

```python
# Exibir as variáveis numéricas mais correlacionadas com a evasão (Churn)
print("\nTop variáveis numéricas mais correlacionadas com Churn:")
churn_corr = corr_matrix['Churn'].drop('Churn').sort_values(key=lambda x: abs(x), ascending=False)
print(churn_corr.head(10))
```

# Análises Direcionadas

```python
# Relação: Tempo de contrato × Evasão (Churn)
if 'Tempo_servico' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Churn', y='Tempo_servico', data=df, palette='Set2', hue= 'Churn')
    plt.title('Tempo de Contrato vs Evasão (Churn)')
    plt.xlabel('Churn')
    plt.ylabel('Tempo de Contrato (meses)')
    plt.xticks([0, 1], ['Não Churn', 'Churn'])
    plt.tight_layout()
    plt.show()
```

```python
# Relação: Total gasto × Evasão (Churn)
if 'Cobranca_total' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Churn', y='Cobranca_total', data=df, palette='Set3', hue= 'Churn')
    plt.title('Total Gasto vs Evasão (Churn)')
    plt.xlabel('Churn')
    plt.ylabel('Total Gasto')
    plt.xticks([0, 1], ['Não Churn', 'Churn'])
    plt.tight_layout()
    plt.show()
```

# Modelagem Preditiva

### Separação dos Dados

```python
from sklearn.model_selection import train_test_split
```

```python
# Separar features e target
X = df_encoded.drop('Churn', axis = 1)
y = df_encoded['Churn']

# Divisão estratificada para manter a proporção de churn

X_escala = X.copy()
X_train, X_test, y_train, y_test = train_test_split(
    X_escala, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

print(f"Shape X_train: {X_train.shape}")
print(f"Shape X_test: {X_test.shape}")
print(f"Shape y_train: {y_train.shape}")
print(f"Shape y_test: {y_test.shape}")

print(f"\nDistribuição do target no conjunto de treino (70%):")
print(y_train.value_counts(normalize=True))

print(f"\nDistribuição do target no conjunto de teste (30%):")
print(y_test.value_counts(normalize=True))
```
